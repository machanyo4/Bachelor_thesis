```python
import numpy as np

eval = np.array([])

for i in range(72):
    start = i*81 + 2
    end = start + 80
    cell_list = worksheet.range("B" + str(start) + ":B" + str(end))
    values = np.array([])
    for cell in cell_list:
        values = np.append(values, int(cell.value))

    sum_val = values.sum()
    values = values / sum_val
    values = np.around(values, decimals = 2)

    eval = np.append(eval, values)

eval = eval.reshape(72, 81)
```

```python
## 0~179までを2進数表記

bin_str = []

for i in range(180):
    bin_str.append(format(i, '08b'))

## char_to_int

bin_list = []

for b_str in bin_str:
    bin_list.append([int(b_int) for b_int in list(b_str)])

# print(bin_list)
```

```python
deg_0 = []
deg_30 = []
deg_60 = []
deg_90 = []
deg_120 = []
deg_150 = []
deg_180 = []
deg_210 = []
deg_240 = []
deg_270 = []
deg_300 = []
deg_330 = []
```

```python
## deg_0 folder

for deg0 in bin_list[0:8]:
    deg_0.append(color_temp[0] + deg0)
for deg0 in bin_list[173:180]:
    deg_0.append(color_temp[0] + deg0)
```

```python
def mk_deg_list(n):
    deg_n = []
    num = (int(n/30)-1)*15+8
    for degn in bin_list[num:num+15]:
        deg_n.append(color_temp[int(n/30)] + degn)
    return(deg_n)
```

```python
## deg_30 folder

deg_30 = mk_deg_list(30)

## deg_60 folder

deg_60 = mk_deg_list(60)

## deg_90 folder

deg_90 = mk_deg_list(90)

## deg_120 folder

deg_120 = mk_deg_list(120)

## deg_150 folder

deg_150 = mk_deg_list(150)

## deg_180 folder

deg_180 = mk_deg_list(180)

## deg_210 folder

deg_210 = mk_deg_list(210)

## deg_240 folder

deg_240 = mk_deg_list(240)

## deg_270 folder

deg_270 = mk_deg_list(270)

## deg_300 folder

deg_300 = mk_deg_list(300)

## deg_330 folder

deg_330 = mk_deg_list(330)
```

```python
deg_list = [deg_0, deg_30, deg_60, deg_90, deg_120, deg_150, deg_180, deg_210, deg_240, deg_270, deg_300, deg_330]
```

```python
def mk_var_tag(temp_list):
    var_tag = []
    for deg in range(len(deg_list)):
        for bin in deg_list[deg]:
            var_tag.append(temp_list + bin)
    return(var_tag)
```

```python
## long_tshits
lt_tag = mk_var_tag(temp_lt)

## long_shirts
ls_tag = mk_var_tag(temp_ls)

## long_poloshirts
lp_tag = mk_var_tag(temp_lp)


## half_tshirts
ht_tag = mk_var_tag(temp_ht)

## half_shirts
hs_tag = mk_var_tag(temp_hs)

## half_poloshirts
hp_tag = mk_var_tag(temp_hp)
```

```python
tag_list = [lt_tag, ls_tag, lp_tag, ht_tag, hs_tag, hp_tag]
```

```python
tops_label = []
for tag in tag_list:
    for label in tag:
        tops_label.append(label)
```

```python
tops_label = np.array(tops_label)
```

```python
train_set = []
test_set = []


for num in range(0,len(tops_label),15):
    train_num = num
    test_num = num+12
    end_num = num+15

    for train_data in tops_label[train_num : test_num]:
        train_set.append(train_data)
    for test_data in tops_label[test_num : end_num]:
        test_set.append(test_data)
```

```python
train_eval = []
test_eval = []


for num in range(0,len(tops_label),15):
    train_num = num
    test_num = num+12
    end_num = num+15

    for train_data in eval_list[train_num : test_num]:
        train_eval.append(train_data)
    for test_data in eval_list[test_num : end_num]:
        test_eval.append(test_data)
```

```python
train = [train_set, train_eval]
test = [test_set, test_eval]
```
